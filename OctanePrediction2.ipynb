```python
samples = pandas.read_csv(os.path.join(datadir, 'FTIR Spectra Sasol blends B1-B14.csv'))
samples.index = samples['wv']
samplesA = 1-samples.ix[:,1:15]
samplesB = samples.ix[:,15:29]
samplesC = samples.ix[:,29:]
samplesA.head()
```

#### Get ranges for samples, normalise, predict using only the scores of the model

```python
A_spec = []

for i in samplesA:
    name = 'Sasol ' + i
    xhold =  sort(array(samplesA.index))
    yhold =  array(samplesA[i])
    firstx = min(xhold) 
    lastx = max(xhold)
    hold = {'CAS REGISTRY NO':name, 'FIRSTX':firstx, 'LASTX':lastx, 'XDATA':xhold, 'YDATA':yhold}
    A_spec.append(hold)
```

```python
B_spec = []

for i in samplesB:
    name = 'Sasol ' + i
    xhold =  sort(array(samplesB.index))
    yhold =  array(samplesB[i])
    firstx = min(xhold) 
    lastx = max(xhold)
    hold = {'CAS REGISTRY NO':name, 'FIRSTX':firstx, 'LASTX':lastx, 'XDATA':xhold, 'YDATA':yhold}
    B_spec.append(hold)
```

```python
C_spec = []

for i in samplesC:
    name = 'Sasol ' + i
    xhold =  sort(array(samplesC.index))
    yhold =  array(samplesC[i])
    firstx = min(xhold) 
    lastx = max(xhold)
    hold = {'CAS REGISTRY NO':name, 'FIRSTX':firstx, 'LASTX':lastx, 'XDATA':xhold, 'YDATA':yhold}
    C_spec.append(hold)
```

```python
A_ranges, A_swn, A_ewn = spec_utils.get_ranges(A_spec)
A_df = spec_utils.spectra_DataFrame(A_spec, A_swn, A_ewn, 4)
A_nf = spec_utils.normalize(A_df)
A_means = spec_utils.mean_data(A_nf)
A_pca, A_myPca = spec_utils.pca_data(A_means.transpose(), 10)
A_pca_df = spec_utils.pca_dataFrame(A_pca, A_spec)
#A_pca_df.head()
```

```python
B_ranges, B_swn, B_ewn = spec_utils.get_ranges(B_spec)
B_df = spec_utils.spectra_DataFrame(B_spec, B_swn, B_ewn, 4)
B_nf = spec_utils.normalize(B_df)
B_means = spec_utils.mean_data(B_nf)
B_pca, B_myPca = spec_utils.pca_data(B_means.transpose(), 10)
B_pca_df = spec_utils.pca_dataFrame(B_pca, B_spec)
#B_pca_df.head()
```

```python
C_ranges, C_swn, C_ewn = spec_utils.get_ranges(C_spec)
C_df = spec_utils.spectra_DataFrame(C_spec, C_swn, C_ewn, 4)
C_nf = spec_utils.normalize(C_df)
C_means = spec_utils.mean_data(C_nf)
C_pca, C_myPca = spec_utils.pca_data(C_means.transpose(), 10)
C_pca_df = spec_utils.pca_dataFrame(C_pca, C_spec)
#C_pca_df.head()
```
