#### First Order Error Plot

```python
errors = []
x_plot = []

for n in linspace(1,30,30):
    n = int(n)
    x_plot.append(n)
    xpca, myPca = spec_utils.pca_data(means.transpose(), n)
    pca_df = spec_utils.pca_dataFrame(xpca, spectra)
    
    full_df = ON.join(pca_df.T)
    
    machine = full_df.ix[:,1:]
    machine.index.name = 'CAS'
    machine.columns = ['C%i' % i for i in machine.columns]
    
    machine = machine.join(prop.ix[:,1:2])
    machine = machine.join(prop.ix[:,3:4])
    
    clf.fit(machine, full_df['RON'])
    
    pre_ron = clf.predict(machine)
    
    avg_err = int(sum(abs(full_df['RON'] - pre_ron)))/len(pre_ron)
    
    
    errors.append(avg_err)
    
plot(x_plot,errors)
xlabel('Number of components')
ylabel('Average Absolute Error')
savefig('abs_error_linear.JPEG')
```

```python
samples = pandas.read_csv(os.path.join(datadir, 'FTIR Spectra Sasol blends B1-B14.csv'))
samples.index = samples['wv']
samplesA = 1-samples.ix[:,1:15]
samplesB = samples.ix[:,15:29]
samplesC = samples.ix[:,29:]
samplesA.head()
```

#### Get ranges for samples, normalise, predict using only the scores of the model

```python
A_spec = []

for i in samplesA:
    name = 'Sasol ' + i
    xhold =  sort(array(samplesA.index))
    yhold =  array(samplesA[i])
    firstx = min(xhold) 
    lastx = max(xhold)
    hold = {'CAS REGISTRY NO':name, 'FIRSTX':firstx, 'LASTX':lastx, 'XDATA':xhold, 'YDATA':yhold}
    A_spec.append(hold)
```

```python
B_spec = []

for i in samplesB:
    name = 'Sasol ' + i
    xhold =  sort(array(samplesB.index))
    yhold =  array(samplesB[i])
    firstx = min(xhold) 
    lastx = max(xhold)
    hold = {'CAS REGISTRY NO':name, 'FIRSTX':firstx, 'LASTX':lastx, 'XDATA':xhold, 'YDATA':yhold}
    B_spec.append(hold)
```

```python
C_spec = []

for i in samplesC:
    name = 'Sasol ' + i
    xhold =  sort(array(samplesC.index))
    yhold =  array(samplesC[i])
    firstx = min(xhold) 
    lastx = max(xhold)
    hold = {'CAS REGISTRY NO':name, 'FIRSTX':firstx, 'LASTX':lastx, 'XDATA':xhold, 'YDATA':yhold}
    C_spec.append(hold)
```
